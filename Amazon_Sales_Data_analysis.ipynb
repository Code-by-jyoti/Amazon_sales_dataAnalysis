```python
#which Country has highest number of sales

df_Country=df.groupby(["Country"])['Total Cost'].sum().sort_values(ascending=False)
df_Country=df_Country.to_frame()
df_Country
```

```python
#What products the most? 

print(df["Item Type"].unique())
print(df["Item Type"].nunique())
```

```python
df_p=df.groupby(['Item Type']) ['Units Sold'].sum().sort_values(ascending=False).head()
df_p=df_p.to_frame()
df_p
```

```python
ItemType=["Cosmetics","Clothes","Beverages","Fruits","Personal Care"]
df_p["ItemType"]=ItemType
```

```python
a4_dims=(11.7,8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(x = "Units Sold",
            y = "ItemType",
            data = df_p)
plt.title("Prouct and Quantity Ordered")
plt.show()
```

```python
sns.histplot(df['Total Revenue'])
plt.xlabel(' Total Revenue') 
plt.ylabel('Density')
```

```python
Order_revenue = df.groupby('Order Date')[['Total Revenue']].sum()
print(Order_revenue)
plt.show()
```

```python
#Determining the best month for the sales

best_month = Order_revenue.sort_values('Total Revenue', ascending=False).index[0] 
print('Best Month:', best_month)
```

```python
#Region with maximum Order

max_region = df.groupby('Region')[['Units Sold']].sum().sort_values('Units Sold', ascending=False).index[0]
print('Region with Max Orders:', max_region) 
plt.show()
```

```python
a4_dims=(20.00,9.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(df['Region'])
plt.xlabel('Region') 
plt.ylabel('Density')
```

```python
#Determining the total pofit 

a4_dims=(11.20,9.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(df['Total Profit'])
plt.xlabel('Total Profit') 
plt.ylabel('Density')
```

```python
import matplotlib.pyplot as plt 
country_names = df.Country.value_counts().index
country_val = df.Country.value_counts().values
# Pie Chart for top 20 country
fig,ax = plt.subplots(figsize=(9,9))
ax.pie(country_val[:20],labels=country_names[:20],autopct='%1.2f%%')
plt.show()
```

```python

import matplotlib.pyplot as plt
sns.set(style='white')
fig, ax = plt.subplots(figsize=(10, 2))
sns.boxplot(df['Total Profit'], color="plum", width=.6)

plt.title('Total Profit Boxplot', fontsize=13)
plt.xlabel('Profit')
plt.show()
     
```
